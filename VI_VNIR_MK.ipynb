# Vegetation Indices

```python

# Load the CSV file
def load_csv(file_path):
    return pd.read_csv(file_path)

# Function to find the closest band for a given wavelength
def get_closest_band(wavelength, wavelenghts):
    closest_index = np.argmin(np.abs(np.array(wavelengths) - wavelength))
    return f'Band_{closest_index + 1}'



# Define functions to calculate each index accurately
def calculate_ari(row, wavelengths):
    R550 = row[get_closest_band(550, wavelengths)]
    R700 = row[get_closest_band(700, wavelengths)]
    return (1 / R550) - (1 / R700)

def calculate_bndvi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R440 = row[get_closest_band(440, wavelengths)]
    R550 = row[get_closest_band(550, wavelengths)]
    return (R800 - R440) / (R800 + R550)

def calculate_bri(row, wavelengths):
    R902 = row[get_closest_band(902, wavelengths)]
    R700 = row[get_closest_band(700, wavelengths)]
    R550 = row[get_closest_band(550, wavelengths)]
    return ((1/R550)-(1/R700)) / R902

def calculate_bwdrvi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R470 = row[get_closest_band(470, wavelengths)]
    return ((0.1 * R800)- R470) / ((0.1 * R800)+ R470)

def calculate_ci(row, wavelengths):
    R683 = row[get_closest_band(683, wavelengths)]
    R690 = row[get_closest_band(690, wavelengths)]
    R675 = row[get_closest_band(675, wavelengths)]
    return (R675*R690 / R683)

def calculate_cig(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    return (R800) / (R670)-1

def calculate_cire(row, wavelengths):
    R760 = row[get_closest_band(760, wavelengths)]
    R720 = row[get_closest_band(720, wavelengths)]
    return (R760 / R720) - 1

def calculate_cri550(row, wavelengths):
    R510 = row[get_closest_band(510, wavelengths)]
    R550 = row[get_closest_band(550, wavelengths)]
    return (1/R510 - 1/R550) 

def calculate_cri700(row, wavelengths):
    R510 = row[get_closest_band(510, wavelengths)]
    R700 = row[get_closest_band(700, wavelengths)]
    return (1/R510 - 1/R700) 

def calculate_cvi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    R550 = row[get_closest_band(550, wavelengths)]
    return R800*(R670/(R550*R550))

def calculate_dd(row, wavelengths):
    R720 = row[get_closest_band(720, wavelengths)]
    R749 = row[get_closest_band(749, wavelengths)]
    R701 = row[get_closest_band(701, wavelengths)]
    R672 = row[get_closest_band(672, wavelengths)]
    return (R749 - R720) / (R701 - R672)

def calculate_dpi(row, wavelengths):
    R688 = row[get_closest_band(688, wavelengths)]
    R710 = row[get_closest_band(710, wavelengths)]
    R697 = row[get_closest_band(697, wavelengths)]
    return (R688 * R710) / (R697 * R697)

def calculate_evi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    R475 = row[get_closest_band(475, wavelengths)]
    return 2.5 * (R800 - R670) / (R800 + 6 * R670 - 7.5 * R475 + 1)

def calculate_exg(row, wavelengths):
    G = row[get_closest_band(550, wavelengths)]
    R = row[get_closest_band(650, wavelengths)]
    B = row[get_closest_band(450, wavelengths)]
    return 2 * G - R - B

def calculate_gli(row, wavelengths):
    G = row[get_closest_band(550, wavelengths)]
    R = row[get_closest_band(670, wavelengths)]
    B = row[get_closest_band(475, wavelengths)]
    return (2 * G - R - B) / (2 * G + R + B)

def calculate_gbndvi(row, wavelengths):
    R860 = row[get_closest_band(860, wavelengths)]
    R530 = row[get_closest_band(530, wavelengths)]
    R440 = row[get_closest_band(440, wavelengths)]
    return (R860 - (R530+R440)) / (R860 + (R530+R440))

def calculate_gosavi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    G = row[get_closest_band(550, wavelengths)]
    return (R800 - G) / (R800 + G + 0.16)

def calculate_grndvi(row, wavelengths):
    R860 = row[get_closest_band(860, wavelengths)]
    R530 = row[get_closest_band(530, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    return (R860 - (R530+R670)) / (R860 + (R530+R670))

def calculate_lci(row, wavelengths):
    R850 = row[get_closest_band(850, wavelengths)]
    R710 = row[get_closest_band(710, wavelengths)]
    R680 = row[get_closest_band(680, wavelengths)]
    return (R850 - R710) / (R850 + R680)

def calculate_mcari(row, wavelengths):
    R700 = row[get_closest_band(700, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    R550 = row[get_closest_band(550, wavelengths)]
    return ((R700 - R670)- 0.2 * (R700 - R550)) * (R700 / R670)

def calculate_mgrvi(row, wavelengths):
    R670 = row[get_closest_band(670, wavelengths)]
    G = row[get_closest_band(550, wavelengths)]
    return ((G*G)-(R670*R670)) / ((G*G)+(R670*R670))

def calculate_mrendvi(row, wavelengths):
    R750 = row[get_closest_band(750, wavelengths)]
    R705 = row[get_closest_band(705, wavelengths)]
    R445 = row[get_closest_band(445, wavelengths)]
    return (R750 - R705) / (R750 + R705 - 2*R445)

def calculate_msavi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    return 0.5 * (2 * R800 + 1 - np.sqrt((2 * R800 + 1) ** 2 - 8 * (R800 - R670)))

def calculate_ndvi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    return (R800 - R670) / (R800 + R670)

def calculate_ndvi_red_edge(row, wavelengths):
    R750 = row[get_closest_band(750, wavelengths)]
    R710 = row[get_closest_band(710, wavelengths)]
    return (R750 - R710) / (R750 + R710)

def calculate_osavi(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    return (R800 - R670) / (R800 + R670 + 0.16)

def calculate_psri(row, wavelengths):
    R678 = row[get_closest_band(678, wavelengths)]
    R500 = row[get_closest_band(500, wavelengths)]
    R750 = row[get_closest_band(750, wavelengths)]
    return (R678 - R500) / R750

def calculate_rdvi1(row, wavelengths):
    R800 = row[get_closest_band(800, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    return (R800 - R670) / np.sqrt(R800 + R670)

def calculate_rgbvi(row, wavelengths):
    G = row[get_closest_band(550, wavelengths)]
    R = row[get_closest_band(650, wavelengths)]
    B = row[get_closest_band(450, wavelengths)]
    return (G**2 - R * B) / (G**2 + R * B)

def calculate_tvi(row, wavelengths):
    R750 = row[get_closest_band(800, wavelengths)]
    R550 = row[get_closest_band(550, wavelengths)]
    R670 = row[get_closest_band(670, wavelengths)]
    return 0.5 * (120 * (R750 - R550) - 200 * (R670 - R550))
```

```python
import pandas as pd

# File path to your CSV file
file_path = 'File_path.csv'

# Load the data
data = pd.read_csv(file_path)

# Select and rename the columns based on the bands
band_mapping = {
    'Band_27': 'B1',
    'Band_53': 'B2',
    'Band_84': 'B3',
    'Band_134': 'B4',
    'Band_153': 'B5',
    'Band_170': 'B6',
    'Band_190': 'B7',
    'Band_214': 'B8',
    'Band_228': 'B8A',
    'Band_264': 'B9'
}

# Select and rename the columns
data = data[list(band_mapping.keys())].rename(columns=band_mapping)

# Define the index calculation functions
def calculate_ndvi1(row):
    return (row['B8'] - row['B4']) / (row['B8'] + row['B4'])

def calculate_ndvi2(row):
    return (row['B8A'] - row['B4']) / (row['B8A'] + row['B4'])

def calculate_ndi45(row):
    return (row['B5'] - row['B4']) / (row['B5'] + row['B4'])
L=0.5
def calculate_savi(row):
    return ((row['B8'] - row['B4']) / (row['B8'] + row['B4'] + L)) * (1.0 + L)

def calculate_GRVI(row):
    return (row['B3'] - row['B4']) / (row['B3'] + row['B4'])

def calculate_dvi(row):
    return row['B8'] - row['B4']

def calculate_ireci(row):
    return (row['B7'] - row['B4']) / (row['B5'] / row['B6'])

def calculate_rvi(row):
    return row['B8'] / row['B4']

def calculate_pssra(row):
    return row['B7'] / row['B4']

def calculate_mcari(row):
    return ((row['B5'] - row['B4']) - 0.2 * (row['B5'] - row['B3'])) * (row['B5'] / row['B4'])
def calculate_ndre(row):
    return (row['B8'] - row['B6']) / (row['B8'] + row['B6'])
def calculate_arvi(row):
    return (row['B8'] - row['B4']-(row['B4'] - row['B2'])) / (row['B8'] + row['B4']-(row['B4'] - row['B2']))

def calculate_gndvi(row):
    return (row['B7'] - row['B3']) / (row['B7'] + row['B3'])

def calculate_ipvi(row):
    return row['B8'] / (row['B8'] + row['B4'])

def calculate_reip(row):
    return 700 + 40 * ((row['B4'] + row['B7']) / 2 - row['B5']) / (row['B6'] - row['B5'])

def calculate_msavi2(row):
    return (2 * row['B8'] - 1 - ((2 * row['B8'] + 1) ** 2 - 8 ) ** 0.5) / 2


# Apply the functions and add the indices to the dataframe
data['NDVI1'] = data.apply(calculate_ndvi1, axis=1)
data['NDVI2'] = data.apply(calculate_ndvi2, axis=1)
data['NDI45'] = data.apply(calculate_ndi45, axis=1)
data['SAVI'] = data.apply(calculate_savi, axis=1)
data['GRVI'] = data.apply(calculate_GRVI, axis=1)
data['DVI'] = data.apply(calculate_dvi, axis=1)
data['IRECI'] = data.apply(calculate_ireci, axis=1)
data['RVI'] = data.apply(calculate_rvi, axis=1)
data['PSSRA'] = data.apply(calculate_pssra, axis=1)
data['MCARI'] = data.apply(calculate_mcari, axis=1)
data['NDRE'] = data.apply(calculate_ndre, axis=1)
data['ARVI'] = data.apply(calculate_arvi, axis=1)
data['GNDVI'] = data.apply(calculate_gndvi, axis=1)
data['IPVI'] = data.apply(calculate_ipvi, axis=1)
data['REIP'] = data.apply(calculate_reip, axis=1)
data['MSAVI2'] = data.apply(calculate_msavi2, axis=1)
# Save the updated DataFrame to a new CSV file
output_file_path = 'file_path.csv'
data.to_csv(output_file_path, index=False)

print(f"File saved at: {output_file_path}")
```
