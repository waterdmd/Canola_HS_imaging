```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load your data
data = pd.read_csv('C:/Users/mkhorram.ASURITE/Desktop/Canola_2/New2/canola_spectrum_index_AND_newindexafterhypertuning.csv')

# Select the features and the target
# X = data[
#     ['Band_180', 'Band_178', 'Band_181', 'Band_179', 'Band_201', 'Band_223',  
#      'RGBVI', 'GLI', 'ExG', 'BWDRVI',   'Band_206', 'Band_257', 
#      'Band_247', 'Band_227', 'Band_222']
# ]
# X = data[
#     ['Band_180', 'Band_178', 'Band_181', 'Band_179', 'Band_201', 'Band_223',  
#      'RGBVI', 'GLI', 'ExG', 'BWDRVI',   'Band_206', 'Band_257', 
#      'Band_247', 'Band_227', 'Band_222']
# ]

X = data[
    ['Band_180', 'Band_178', 'Band_181', 'Band_179', 'Band_201', 'Band_223',  
     'NEW_index','RGBVI','GLI','ExG','BWDRVI','CVI','BWDRVI','CIG','CRI700','DPI','DD',   'Band_206', 'Band_257', 
     'Band_247', 'Band_227', 'Band_222']
]

y = data['salinity']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7641)

# Initialize and train the RidgeClassifier
model = RidgeClassifier(alpha=0.12648552168552957, tol=0.1, solver='cholesky', random_state=7641)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate overall accuracy
initial_accuracy = accuracy_score(y_test, y_pred)
print(f"Initial Model Accuracy on Test Set: {initial_accuracy:.4f}")

# Show classification report with per-class accuracy
print("\nClassification Report (Per-Class Accuracy):")
print(classification_report(y_test, y_pred))

# Display the classes in the dataset
print(f"\nClasses in the dataset: {np.unique(y)}")

# Show the first few rows of test results with their corresponding predicted class
test_results = pd.DataFrame({
    'True Class': y_test,
    'Predicted Class': y_pred
}).head()

print("\nHead of Test Results:")
print(test_results)

# Feature importance calculation
feature_importance = np.mean(np.abs(model.coef_), axis=0)

# Create a dataframe for feature importance
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
})

# Sort by importance and get the top 20 features
top_features = importance_df.sort_values(by='Importance', ascending=False).head(20)

# Plot the top 20 feature importances
plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'], color='green')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Top 20 Feature Importances')
plt.gca().invert_yaxis()  # To have the most important feature on top
plt.show()

# Perform feature elimination to find the best feature to remove
best_accuracy = initial_accuracy
best_feature_to_remove = None

for feature in X.columns:
    # Create a temporary DataFrame without the current feature
    X_temp = X.drop(columns=[feature])
    
    # Split the data into training and testing sets again
    X_train_temp, X_test_temp, y_train_temp, y_test_temp = train_test_split(X_temp, y, test_size=0.2, random_state=7641)
    
    # Train the model
    model.fit(X_train_temp, y_train_temp)
    
    # Predict on the test set
    y_pred_temp = model.predict(X_test_temp)
    
    # Calculate accuracy
    temp_accuracy = accuracy_score(y_test_temp, y_pred_temp)
    
    # Check if the accuracy improves by removing the feature
    if temp_accuracy >= best_accuracy:
        best_accuracy = temp_accuracy
        best_feature_to_remove = feature
        print(f"New Best Accuracy: {best_accuracy:.4f} by removing feature: {feature}")

# Output the results
if best_feature_to_remove:
    print(f"\nBest Accuracy achieved by removing: {best_feature_to_remove} with an accuracy of {best_accuracy:.4f}")
else:
    print("\nNo improvement in accuracy by removing any feature.")
```

#  VI Selection

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import accuracy_score
import warnings

warnings.filterwarnings('ignore')

# Load your data
data = pd.read_csv('C:/Users/mkhorram.ASURITE/Desktop/Canola_2/New2/canola_spectrum_index_AND_newindexafterhypertuning.csv')

# Identify the "Band" columns and the columns to exclude ('salinity' and 'variety')
band_columns = [col for col in data.columns if col.startswith('Band')]
exclude_columns = ['salinity', 'variety']

# Select the remaining input features (excluding 'salinity' and 'variety')
remaining_features = data.drop(columns=exclude_columns + band_columns).columns

# Target variable
y = data['salinity']

# Initialize lists to store feature names and their corresponding accuracies and best parameters
features = []
accuracies = []
best_params_list = []

# Function to perform random search and hypertuning for a given set of features
def hypertune_model(X, y, feature_name=None):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7641)
    
    # Define the parameter grid for random search
    param_distributions = {
        'alpha': np.logspace(-4, 4, 50),  # Ridge regularization strength
        'tol': [1e-4, 1e-3, 1e-2, 1e-1, 1],  # Tolerance for stopping criteria
        'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'lbfgs'],  # Different solvers
    }
    
    # Perform Randomized Search with cross-validation
    random_search = RandomizedSearchCV(
        estimator=RidgeClassifier(random_state=7641),
        param_distributions=param_distributions,
        n_iter=100,  # Number of random configurations to sample
        cv=5,  # 5-fold cross-validation
        scoring='accuracy',  # Evaluation metric
        n_jobs=-1,  # Use all available cores
        random_state=7641,  # For reproducibility
        verbose=0  # Suppress progress
    )
    
    # Fit the model on the training data
    random_search.fit(X_train, y_train)
    
    # Get the best model and parameters
    best_model = random_search.best_estimator_
    best_params = random_search.best_params_
    
    # Predict on the test set
    y_pred = best_model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Store results
    features.append(feature_name)
    accuracies.append(accuracy)
    best_params_list.append(best_params)

    print(f"Feature: {feature_name}, Accuracy: {accuracy:.4f}, Best Params: {best_params}")

# First iteration: Use only the "Band" columns (without adding any additional features)
X_bands_only = data[band_columns]
hypertune_model(X_bands_only, y, feature_name='Only_Bands')

# Iterate through each remaining feature and add one at a time to the band columns
for feature in remaining_features:
    # Use all "Band" columns plus one additional feature at a time
    X = pd.concat([data[band_columns], data[[feature]]], axis=1)
    hypertune_model(X, y, feature_name=feature)

# Create a DataFrame to store features, accuracies, and best parameters
results_df = pd.DataFrame({
    'Feature': features,
    'Accuracy': accuracies,
    'Best Parameters': best_params_list
})

# Sort the DataFrame by accuracy in descending order
sorted_results_df = results_df.sort_values(by='Accuracy', ascending=False)

# Save the sorted results to a CSV file
sorted_results_df.to_csv('C:/Users/mkhorram.ASURITE/Desktop/Canola_2/New2/feature_accuracy_ranking2with_bands.csv', index=False)

# Display the sorted results
print(sorted_results_df)
```
